```python
from langchain_openai import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.schema import Document
```

```python
doc1 = Document(
    page_content="3 Idiots is a blockbuster Bollywood film known for its strong message on education.",
    metadata={"genre": "drama"}
)
doc2 = Document(
    page_content="Dangal is a sports biographical film based on the life of wrestler Mahavir Singh Phogat.",
    metadata={"genre": "sports"}
)
doc3 = Document(
    page_content="Chennai Express is a Bollywood action-comedy film starring Shah Rukh Khan and Deepika Padukone.",
    metadata={"genre": "action-comedy"}
)
doc4 = Document(
    page_content="Kabir Singh is a romantic drama film starring Shahid Kapoor and Kiara Advani.",
    metadata={"genre": "romance"}
)

doc5 = Document(
    page_content="Bahubali is a historical action movie known for its grand visuals and storytelling.",
    metadata={"genre": "action"}
)
```

```python
docs = [doc1,doc2,doc3,doc4,doc5]
```

```python
vector_store = Chroma(embedding_function=OpenAIEmbeddings(model="text-embedding-3-small",api_key=""),
                      persist_directory="my_chroma_bollywood_db",
                      )
```

```python
vector_store.add_documents(docs)
```

```python
vector_store.get(include=['embeddings','documents','metadatas'])sss
```

```python
vector_store.similarity_search(query="which movie is about sports?",
                               k = 2, # returns number of neariest neighbour
                               )
```

```python
vector_store.similarity_search_with_score(
    query = "Which movie is based on education?",
    k = 2,
)
```

```python
vector_store.similarity_search_with_score(
    query="action movie",
    k=3,
    filter={"genre":"action"}
)
```

```python
# A Document
update_doc1 = Document(
    page_content="3 Idiots is a blockbuster Bollywood film known for its strong message on education.",
    metadata={"genre": "horror"},
)
vector_store.update_document(
    document_id = "d4a9a5b7-38ed-4b4b-ac17-b17c7dd0034d",
    document=update_doc1,
)
```
